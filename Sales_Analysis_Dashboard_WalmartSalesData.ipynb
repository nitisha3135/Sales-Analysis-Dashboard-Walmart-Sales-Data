## üìù Steps to Build the Project 

### üîπ Step 1: Load the Dataset

First, let's load the dataset and check for missing values.

```python
import pandas as pd

# Load dataset
df = pd.read_csv(r"D:\Data Analytics Projects\Project 1\archive\Walmart.csv")

# Display the first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())
```

-----

### üîπ Step 2: Convert Data Types

We need to ensure the Date column is in datetime format.

```python
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
```

-----

### üîπ Step 3: Sales Trend Over Time

We will visualize the total sales over time.

```python
import matplotlib.pyplot as plt

# Group sales by Date
sales_trend = df.groupby("Date")["Weekly_Sales"].sum()

# Plot sales trend
plt.figure(figsize=(12, 6))
plt.plot(sales_trend.index, sales_trend.values, marker="o", linestyle="-")
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

-----

### üîπ Step 4: Sales Impact During Holidays

Let's analyze whether holidays impact sales.

```python
import seaborn as sns

# Boxplot of sales on holiday vs non-holiday weeks
plt.figure(figsize=(8, 6))
sns.boxplot(x="Holiday_Flag", y="Weekly_Sales", data=df)
plt.xticks([0, 1], ["Non-Holiday", "Holiday"])
plt.title("Sales Distribution: Holiday vs Non-Holiday")
plt.show()
```

-----

### üîπ Step 5: Store-Wise Sales Performance

Identify top-performing stores.

```python
# Group by store and sum sales
store_sales = df.groupby("Store")["Weekly_Sales"].sum().sort_values(ascending=False)

# Bar plot of top 10 stores
plt.figure(figsize=(10, 6))
store_sales.head(10).plot(kind="bar", color="blue")
plt.title("Top 10 Best-Performing Stores")
plt.xlabel("Store")
plt.ylabel("Total Sales")
plt.show()
```

------

### üîπ Step 6: Correlation Analysis

Find correlations between sales and economic factors.

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

------

### üîπ Step 7: Predict Future Sales Using Linear Regression

We'll build a simple regression model.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

# Selecting features and target
features = ["Temperature", "Fuel_Price", "CPI", "Unemployment"]
target = "Weekly_Sales"

# Split data
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
```
